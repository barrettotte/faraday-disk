# Faraday Disk

A brief exploration of the theoretical performance of Faraday disks.

Induced EMF for Faraday Disk (from center to rim):

$$\epsilon=\frac{1}{2}\omega BR^2$$
where
- $\omega$ is the angular velocity of the disk in rad/s
- $B$ is the magnetic flux density in Teslas
- $R$ is the radius of the disk in meters

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# calculate emf using formula in first cell
def induced_emf(omega, B, R):
    return 0.5 * omega * B * R**2

# convert RPM to rad/s
def rpm_to_rps(rpm):
    return (2 * np.pi * rpm) / 60
```

```python
# varying B and omega, with static R

omega_values = np.linspace(0, rpm_to_rps(5000), 100)
B_values = np.linspace(0, 2.0, 100)
R = 0.15

omega, B = np.meshgrid(omega_values, B_values)
emf = induced_emf(omega, B, R)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(omega, B, emf, cmap='viridis')
ax.set_box_aspect(aspect=None, zoom=0.9)

ax.set_xlabel('Angular Velocity (rad/s)')
ax.set_ylabel('Magnetic Flux Density (T)')
ax.set_zlabel('Induced EMF (V)')

plt.title(f'Induced EMF (Radius={R})')
plt.tight_layout()
plt.show()
```
